```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
ab = pd.read_csv('../Datasets/abalone.csv')
```

```python
ab
```

```python
ab.dtypes
ab.loc[ab['Sex']=='M','Sex']=0
ab.loc[ab['Sex']=='F','Sex']=1
ab.loc[ab['Sex']=='I','Sex']=2
```

```python
X = ab.drop('Rings',axis=1).values
y = ab['Rings'].values
```

```python
X_Length = X[:,4]
type(X_Length), type(y)
print(X_Length)
```

```python
y = y.reshape(-1,1)
```

```python
X_Length = X_Length.reshape(-1,1)
```

```python
%config InlineBackend.figure_format = 'svg'
plt.scatter(X_Length, y)
plt.ylabel('Age')
plt.xlabel('Shell Length')
plt.show();
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit(X_Length,y)
prediction_space=np.linspace(min(X_Length),max(X_Length)).reshape(-1,1)
```

```python
plt.scatter(X_Length,y,color='red')
plt.plot(prediction_space, reg.predict(prediction_space), color='black', linewidth=3)
plt.ylabel('Age')
plt.xlabel('Shell Length')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
reg_all = linear_model.LinearRegression()
reg_all.fit(X_train, y_train)
y_pred = reg_all.predict(X_test)
reg_all.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
reg = linear_model.LinearRegression()
cv_results = cross_val_score(reg,X,y,cv=3)
print(cv_results)
```
