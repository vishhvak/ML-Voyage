```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn import model_selection
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
#for not being able to get a dataset which is difficult to classify, I'm using sklearns make_classification function 
# to generate an artificial dataset with less separation between the classes
X, y = make_classification(n_samples=1000, 
    n_features=5, 
    n_redundant = 0,
    n_informative = 3,
    n_repeated = 0,
    n_classes = 3,
    n_clusters_per_class = 1,
    class_sep=1.75, 
    flip_y=0.1,
    random_state=1) 
```

```python
y
```

```python
df = pd.DataFrame({'x1':X[:,0], 'x2':X[:,1], 'x3':X[:,2], 'x4':X[:,3], 'y':y})
df
```

```python
df.head()
```

```python
import seaborn as sb
%config InlineBackend.figure_format = 'svg'
#Correlation heatmap for dataset
heat = sb.heatmap(df.corr(),annot=True, fmt='.2f')
```

```python
#Visualizing Dataset
print(X)
sb.pairplot(df, height=3,vars=['x1','x2','x3','x4'],hue="y",palette="husl")
```

```python
#Train and Test Data Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix")
from sklearn import tree
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
d_pred = dtc.predict(X_test)
print("Confusion Matrix")
print(confusion_matrix(y_test, d_pred))
print('\n')
print("Classification Report")
print(classification_report(y_test, d_pred))
print('\n')
accuracy = accuracy_score(y_test,d_pred)
print("Accuracy:%.2f%%" %(accuracy*100.0))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rpredict = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix")
print(confusion_matrix(y_test, rpredict))
print('\n')
print("Classification Report")
print(classification_report(y_test, rpredict))
print('\n')
accuracy = accuracy_score(y_test,rpredict)
print("Accuracy:%.2f%%" %(accuracy*100.0))
```

```python
#Binary classification for Adaboost
X, y = make_classification(n_samples=1000, 
    n_features=8, 
    n_redundant = 2,
    n_informative = 4,
    n_repeated = 2,
    n_classes = 2,
    n_clusters_per_class = 2,
    class_sep=0.75, 
    flip_y=0.001,
    random_state=1) 
```
